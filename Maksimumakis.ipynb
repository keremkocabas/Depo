```python
import networkx as nx
import matplotlib.pyplot as plt

# Grafı oluşturuyoruz
G = nx.DiGraph()  # Yönlü grafik (Direct Graph) kullanıyoruz çünkü akış yönlüdür

# Düğüm ve kenarları ekliyoruz. Kenarların kapasitelerini belirtiyoruz.
G.add_edge('A', 'B', capacity=4)
G.add_edge('A', 'C', capacity=2)
G.add_edge('B', 'C', capacity=5)
G.add_edge('B', 'D', capacity=10)
G.add_edge('C', 'D', capacity=3)
G.add_edge('D', 'E', capacity=4)
G.add_edge('D', 'F', capacity=6)
G.add_edge('E', 'F', capacity=1)


# Maksimum akışı hesaplıyoruz (Edmonds-Karp algoritması kullanılıyor)
flow_value, flow_dict = nx.maximum_flow(G, 'A', 'F')

# Sonuçları yazdırıyoruz
print("Maksimum Akış Değeri:", flow_value)
print("Akışın Dağılımı:")
for node in flow_dict:
    print(f"{node}: {flow_dict[node]}")

# Grafı çiziyoruz
pos = nx.spring_layout(G)  # Düğümlerin yerleşimini belirler
plt.figure(figsize=(8, 6))

# Grafı çiziyoruz
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=15, font_weight="bold", edge_color='gray')

# Kenarların kapasite etiketlerini ekliyoruz
edge_labels = nx.get_edge_attributes(G, 'capacity')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Grafı gösteriyoruz
plt.title("Graf ve Maksimum Akış (A'dan F'ye)")
plt.show()
```
